```python
# zero_elimination_map.py
"""
Visualize the regional elimination strategy for the Riemann Hypothesis.
This plot shows excluded regions in the complex plane and highlights
the critical line Re(s) = 1/2 as the only constructively valid zone for nontrivial zeros.
"""

import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots(figsize=(10, 6))
ax.set_xlim(-1, 2)
ax.set_ylim(-20, 20)
ax.axvline(0, color='black', linewidth=1)
ax.axhline(0, color='black', linewidth=1)
ax.set_xlabel('Re(s)')
ax.set_ylabel('Im(s)')
ax.set_title('Zero Elimination Map for Riemann Zeta Function')

regions = [
    (-1.0, 0.0, 'lightgrey', 'Excluded (Divergence)'),
    (0.0, 0.5, 'lightcoral', 'Excluded (Symmetry)'),
    (0.5, 1.0, 'lightsalmon', 'Excluded (Interference asymmetry)'),
    (1.0, 1.01, 'lightblue', 'Excluded (Log stability)'),
    (1.01, 2.0, 'lightgreen', 'Excluded (Absolute convergence)'),
]

for x_start, x_end, color, label in regions:
    ax.add_patch(patches.Rectangle((x_start, -20), x_end - x_start, 40,
                                   color=color, label=label))

ax.axvline(0.5, color='blue', linewidth=2, linestyle='--', label='Critical Line: Re(s) = 1/2')

arrow_props = dict(arrowstyle='->', color='black')
ax.annotate('', xy=(-0.2, 0), xytext=(0.4, 0), arrowprops=arrow_props)
ax.annotate('', xy=(1.3, 0), xytext=(0.6, 0), arrowprops=arrow_props)
ax.text(0.6, 1, 'Elimination flow', fontsize=10)

ax.text(-0.5, 18, 'Region C.5', fontsize=9)
ax.text(0.25, 18, 'Region C.4', fontsize=9)
ax.text(0.75, 18, 'Region C.3', fontsize=9)
ax.text(1.0, 18, 'Region C.2', fontsize=9)
ax.text(1.5, 18, 'Region C.1', fontsize=9)
ax.text(0.52, -18, '← Only viable zone (critical line)', fontsize=10, color='blue')

handles, labels = ax.get_legend_handles_labels()
by_label = dict(zip(labels, handles))
ax.legend(by_label.values(), by_label.keys(), loc='upper right')

plt.grid(True)
plt.tight_layout()
plt.savefig('__appendix_C_zero_elimination_map.png', dpi=300)  # Save the plot
plt.show()
plt.close()
```
