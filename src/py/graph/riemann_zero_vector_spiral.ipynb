### 幾何学的ゼロ点構造の視覚化

この図は、ゼータ関数の非自明ゼロ点付近 (Re(s)=1/2, t≈14.135) における、<br />
複素ベクトル $\frac{1}{n^{s}} e^{-it \log n}$ の和を描いたものです。

ベクトルが巻きながらゼロへ向かうことは、「調和的位相干渉」による<br />
“打ち消し”の構造であり、これがゼロ点を構成する仮説的根拠となる視点を示しています。


```python
import numpy as np
import matplotlib.pyplot as plt

# 日本語フォント設定
plt.rcParams['font.family'] = 'kiloji'

# パラメータ設定
s_real = 0.5  # Re(s) = 1/2 固定
t = 14.135  # 最初の非自明なゼロ点の近傍
n_max = 500  # ベクトル数
n = np.arange(1, n_max + 1)

# ベクトル生成：1/n^s * exp(-it log n)
r = 1 / (n ** s_real)
theta = -t * np.log(n)
vectors = r * np.exp(1j * theta)

# ベクトルの和を累積
points = np.cumsum(vectors)

# プロット準備
fig, ax = plt.subplots(figsize=(8, 8))
ax.plot(points.real, points.imag, lw=1.2, label=f"∑ 1/n^s · e^(-it log n), t = {t}")
ax.quiver(points[:-1].real, points[:-1].imag,
          np.diff(points).real, np.diff(points).imag,
          angles='xy', scale_units='xy', scale=1, width=0.002, alpha=0.4)

# 始点・終点を目立たせる
ax.plot([0], [0], 'go', label="Start (0, 0)")
ax.plot([points[-1].real], [points[-1].imag], 'ro', label="End")

ax.set_title("ベクトル和によるゼロ点構造（螺旋の打ち消し）")
ax.set_xlabel("Re")
ax.set_ylabel("Im")
ax.grid(True)
ax.axis('equal')
ax.legend()

# 保存
# image_path = "riemann_zero_vector_spiral.png"
# plt.savefig(image_path, dpi=300)
plt.show()
plt.close()

# image_path
```
